```python
import pandas as pd

# Assuming you have a DataFrame 'mcdonalds' with the relevant columns
# Select the columns you want to include in MD.x
selected_columns = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting']

# Create a binary matrix where 1 represents "Yes" and 0 represents "No"
MD_x = (df_mcd[selected_columns] == 'Yes').astype(int)

# Calculate column means and round to 2 decimal places
column_means = MD_x.mean().round(2)

# Display the result
print(column_means)
```

```python
from sklearn.decomposition import PCA
import pandas as pd

# Assuming you have the binary matrix MD_x as previously defined

# Perform PCA
pca = PCA()
MD_pca = pca.fit(MD_x)

# Get the standard deviations, proportion of variance, and cumulative proportion
standard_deviations = MD_pca.explained_variance_**0.5
proportion_of_variance = MD_pca.explained_variance_ratio_
cumulative_proportion = proportion_of_variance.cumsum()

# Create a summary DataFrame
summary_df = pd.DataFrame({
    'Standard Deviation': standard_deviations,
    'Proportion of Variance': proportion_of_variance,
    'Cumulative Proportion': cumulative_proportion
}, index=[f'PC{i+1}' for i in range(len(standard_deviations))])

# Print the summary
print(summary_df)
```

```python
from sklearn.decomposition import PCA
import pandas as pd

# Assuming you have the binary matrix MD_x as previously defined

# Perform PCA
pca = PCA()
MD_pca = pca.fit(MD_x)

# Print the standard deviations (eigenvalues)
print("Standard deviations (1, .., p=11):")
print(pca.explained_variance_**0.5)

# Print the rotation matrix (loading scores)
print("Rotation (n x k) = (11 x 11):")
rotation_matrix = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(len(pca.components_))], index=selected_columns)
print(rotation_matrix.round(3))
```

```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Assuming you have the binary matrix MD_x as previously defined

# Perform PCA
pca = PCA()
MD_pca = pca.fit(MD_x)

# Plot consumers projected onto the first two principal components (in grey)
projected_data = pca.transform(MD_x)
plt.scatter(projected_data[:, 0], projected_data[:, 1], color="grey", label="Consumers")

# Plot arrows for original segmentation variables
for i, var in enumerate(selected_columns):
    plt.arrow(0, 0, pca.components_[0, i], pca.components_[1, i], color='red', alpha=0.5, width=0.01)
    plt.text(pca.components_[0, i], pca.components_[1, i], var, color='red', fontsize=12)

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Projection of Data')
plt.legend()
plt.grid(True)

plt.show()
```
